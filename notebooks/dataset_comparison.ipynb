```python
def generate_subplots(rows, cols, paths) -> None:
    assert len(paths)==rows*cols, "Not enough images!"
    fig, axs = plt.subplots(rows, cols, figsize=(15, 11))
    fig.tight_layout(h_pad=1.2, w_pad=.5)
    row, col = 0, 0
    for path in paths:
        if col == cols:
            row += 1
            col = 0
        img = Image.open(path)
        arr = np.asarray(img)
        axs[row, col].imshow(img, cmap=plt.cm.bone)
        name = "min: " + str(np.around(arr.min(), 2)) + ", max: " + str(np.around(arr.min(), 2)) + ", mean: " + str(np.around(arr.mean(), 2))
        axs[row, col].set_title(name)
        col += 1
```

```python
generate_subplots(5, 5, paths[:25])
```
