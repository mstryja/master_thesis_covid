```python
import tensorflow as tf
from master_lib.utils import files_and_directories
from master_lib.image_processing.processing import generate_numpy_dataset
```

```python
model = tf.keras.models.load_model(r'C:\Users\MikołajStryja\Documents\Studia\master\trained_models\Sequential_4')
```

```python
model.summary()
```

```python
dataset_path = r'C:\Users\MikołajStryja\Documents\Studia\master\dataset.json'
dataset = files_and_directories.load_json(dataset_path)
dataset_images = r'C:\Users\MikołajStryja\Documents\Studia\siim-covid19-detection'
dataset = files_and_directories.load_json(dataset_path)
train_imgs = files_and_directories.list_train_images(dataset_images)
```

```python
evaluation = generate_numpy_dataset(train_imgs[-100:], dataset, division_coef=0)
evaluation
```

```python
X, Y = evaluation[0], evaluation[1]
```

```python
Y.shape
```

```python
results = model.predict(X)
```

```python
import numpy as np
for elem in results:
    max_value = np.where(elem == np.amax(elem))
    elem[max_value] = 1
    elem[elem != 1] = 0
```

```python
for x, y in zip(results, Y):
    print(f"Predicted: {x}, True label: {y}")
```

```python
m = tf.keras.metrics.Precision()
r = tf.keras.metrics.Recall()
m.update_state(results, Y)
r.update_state(results, Y)
precision = m.result().numpy()
recall = r.result().numpy()
precision, recall, 2*precision*recall/(precision + recall)
```
